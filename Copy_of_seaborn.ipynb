```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://github.com/wizardMati/matplotlip/blob/main/HRDataset.csv")

df["DateofHire"] = pd.to_datetime(df["DateofHire"])
df["DateofTermination"] = pd.to_datetime(df["DateofTermination"])
df["DOB"] = pd.to_datetime(df["DOB"])

df["EndDate"] = df["DateofTermination"].fillna(pd.Timestamp.today())
df["TenureYears"] = (df["EndDate"] - df["DateofHire"]).dt.days / 365
df["Age"] = (pd.Timestamp.today() - df["DOB"]).dt.days // 365

sns.set_theme(style="whitegrid")

sns.catplot(
    data=df,
    x="ManagerName",
    hue="PerformanceScore",
    kind="count",
    height=6,
    aspect=2
).set(
    title="Performance Score by Manager",
    xlabel="Manager",
    ylabel="Number of Employees"
)

sns.catplot(
    data=df,
    x="RecruitmentSource",
    y="TenureYears",
    kind="bar",
    estimator="mean",
    height=6,
    aspect=2
).set(
    title="Average Tenure by Recruitment Source",
    xlabel="Recruitment Source",
    ylabel="Average Tenure (Years)"
)

sns.catplot(
    data=df,
    x="MaritalDesc",
    y="EmpSatisfaction",
    kind="box",
    height=5,
    aspect=1.5
).set(
    title="Employee Satisfaction by Marital Status",
    xlabel="Marital Status",
    ylabel="Employee Satisfaction"
)

sns.displot(
    data=df,
    x="Age",
    bins=15,
    kde=True,
    height=5,
    aspect=1.5
).set(
    title="Age Distribution of Employees",
    xlabel="Age",
    ylabel="Number of Employees"
)

sns.lmplot(
    data=df,
    x="Age",
    y="SpecialProjectsCount",
    height=5,
    aspect=1.5
).set(
    title="Age vs Number of Special Projects",
    xlabel="Age",
    ylabel="Number of Special Projects"
)
```
